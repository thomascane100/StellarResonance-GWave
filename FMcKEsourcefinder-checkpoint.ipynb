```python
import pandas as pd
from math import *
from scipy import interpolate
import matplotlib.pyplot as plt
import numpy as np
from decimal import *
getcontext().prec=20

#import data
cosmicpop = pd.read_csv(r"YourPathway", index_col=False)

#define values from df
porb=cosmicpop.OrbitalPeriod
mass_1Solar=cosmicpop.mass1
mass_2Solar=cosmicpop.mass2
mass_1rSolar=cosmicpop.mass1r
mass_2rSolar=cosmicpop.mass2r
a_b=cosmicpop.SemiMajorAxis
#ecc=cosmicpop.ecc
mc=cosmicpop.Chirpmass
w_star1=cosmicpop.wstar1
w_star2=cosmicpop.wstar2
ecc.cosmicpop.ecc
#constants
#gravitational converted from AU**3*yr**-2*Msolar**-1 to Solar Radii*day**-2*Msolar**-1
G=3100670.345563776
#speed of light in solar radii per day
c=37231.7

#define roche lobe radii
q_1=mass_1Solar/mass_2Solar
r_l_1=(0.49*q_1**(2/3))/(0.6*q_1**(2/3)+np.log(1+q_1**(1/3)))
q_2=mass_2Solar/mass_1Solar
r_l_2=(0.49*q_2**(2/3))/(0.6*q_2**(2/3)+np.log(1+q_2**(1/3)))
r_l_tot=r_l_1+r_l_2
#reduced mass
mu=(mass_1Solar*mass_2Solar)/(mass_1Solar+mass_2Solar)

#break up spin fraction 
fspin1=-1
w_spin1=fspin1*w_star1


#principal quantum numbers
q=1
m=2
#harmonic counter
k=2
#forcing frequency
w_force=q*porb-m*w_spin1

#initial decay time (days)
tdp=11574074.074074075
tdg=11574074.074074075

#find the resonnat limits to high precision
for i in range(len(cosmicpop)):
    newhi=Decimal(w_star1[i]/k+1/tdp)
    newlo=Decimal(w_star1[i]/k-1/tdp)
    
    
#initial mode coefficients
d_res_mode_p=np.float32(2)
d_res_mode_g=0.1

#forcing frequency range by varying break spin fraction
rangeforce=[]
breakup=[]
for i in range(len(cosmicpop)):
    while fspin1<=1:
        w_spin1=fspin1*w_star1
        w_force=q*porb-m*w_spin1
        fspin1+=0.01
        rangeforce.append(w_force)
        breakup.append(fspin1)

#flatten list of lists
spinbreakrange=list(zip(*rangeforce))

newforcerange=[item for sublist in spinbreakrange for item in sublist]

#source meets resonance criteria
resonantsource=[]
breakfrac=[]
for i in range(len(cosmicpop)):
    if (Decimal(newlo)-Decimal(1*10**-5)) <=newforcerange[i] and newforcerange[i]<=(Decimal(newhi)+Decimal(1*10**-5)):
        resonantsource.append(newforcerange[i])
        breakfrac.append(breakup[i])

modecoeffg=[]
modecoeffp=[]
#find the type of mode by finding the coefficient using division
for i in range(len(resonantsource)):
    starcoeff=resonantsource[i]/w_star1[i]
    #pmode
    if starcoeff>=2:
        modecoeffp.append(starcoeff)
    #gmode
    elif starcoeff<=1:
        modecoeffg.append(starcoeff)
```
