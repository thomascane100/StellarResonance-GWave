```python
import pandas as pd
from math import *
from scipy import interpolate
import matplotlib.pyplot as plt
import numpy as np
from decimal import *
```

```python
#import data
cosmicpop = pd.read_csv(r"PathToFile", index_col=False)

#define values from df
PorbSI=cosmicpop.PorbSI
mass_1SI=cosmicpop.mass_1SI
mass_2SI=cosmicpop.mass_2SI
mass_1rSI=cosmicpop.mass_1rSI
mass_2rSI=cosmicpop.mass_2rSI
a_b=cosmicpop.a_b
ecc=cosmicpop.ecc
mc=cosmicpop.mc
w_star1=cosmicpop.w_star1
w_star2=cosmicpop.w_star2
```

```python
#re convert values for sensitivity
############
PorbDay=PorbSI/86400
#PorbYear=porb/356.256

#convert mass1/2 from kg to solar mass
mass_1Solar=mass_1SI/1.98847e30
mass_2Solar=mass_2SI/1.98847e30
#convert m to solar radii
mass_1rSolar=mass_1rSI/6.96e8
mass_2rSolar=mass_2rSI/6.96e8
#chirp mass
mc=((mass_1Solar*mass_2Solar)**(3/5)/((mass_1Solar+mass_2Solar)**(1/5)))
#semi-major from meters to solar radii
a_b=a_b/6.96e8
#angular velocity from s^-1 to yr^-1
w_star1=w_star1/86400
w_star2=w_star2/86400
```

```python
#constants
#gravitational converted from AU**3*yr**-2*Msolar**-1 to Solar Radii*day**-2*Msolar**-1
G=3100670.345563776
#speed of light in solar radii per day
c=37231.7
p=0
#define roche lobe radii
q_1=mass_1Solar/mass_2Solar
r_l_1=(0.49*q_1**(2/3))/(0.6*q_1**(2/3)+np.log(1+q_1**(1/3)))
q_2=mass_2Solar/mass_1Solar
r_l_2=(0.49*q_2**(2/3))/(0.6*q_2**(2/3)+np.log(1+q_2**(1/3)))
r_l_tot=r_l_1+r_l_2
#reduced mass
mu=(mass_1Solar*mass_2Solar)/(mass_1Solar+mass_2Solar)
#times
#t=0
tobs=1461.03
t_gw=(5/128)*((c**5)/(G**3))*((a_b**4)/((mass_1Solar+mass_2Solar)**2*mu))*(1-ecc**2)**(7/2)
```

```python
#variables for gwave evolution
Beta=(64/5)*c**(-5)*G**3*mass_1Solar*mass_2Solar*(mass_1Solar+mass_2Solar)

a0=a_b

c0=a_b*(1-ecc**2)*ecc**(-12/19)*(1+(121/304)*ecc**2)**(-870/2299)

cn=a_b*(1-ecc**2)*ecc**(-12/19)*(1+(121/304)*ecc**2)**(-870/2299)

porb0=(a0**3/(mass_1Solar+mass_2Solar))**(1/2)

porb=(a_b**3/(mass_1Solar+mass_2Solar))**(1/2)


freq0=2/porb0

E_GW=0

Eorb=-(G*(mass_1Solar+mass_2Solar))/(2*a_b)

eorbtot=Eorb+E_GW
```

```python
#principal quantum numbers
q=1
m=2
#
#forcing frequency
w_force=q*porb0-m*w_spin1
#mode counter
d_res_mode_g=0.1
d_res_mode_p=2
#energy of WD and NS
Estar_1=(G*mass_1Solar**2)/mass_1rSolar
Estar_2=(G*mass_2Solar**2)/mass_2rSolar
Esys=Estar_1+Estar_2
#initial mode test
pmode0=d_res_mode_p*w_star1
gmode0=d_res_mode_g*w_star1
```

```python
#harmonic counter
k=2
r=1
#typical decay time (days)
t_p=0.009296076592592595
#number of radial modes
Np=10
#initial decay time of mode (p or g) measured in days
tdp=11574074.074074075
tdg=11574074.074074075
#initial energy deposited into modes
Ep0=0
Eg0=0
Ep=Ep0
Eg=Eg0
#total resonant energy
Ert0=Ep0+Eg0
Ert=Ep+Eg
#initial total energy
#Etotal0=Ert0+Esys+Eorb0+E_GW0
```

```python
#FMcKE resonant region stipulations
low_end=w_star1/k-1/tdp
high_end=w_star1/k+1/tdp
```

```python
#grav wave lists
orbital_period=[]
Grav_wave_energy=[]
semi_major=[]
frequency=[]
newecc=[]
#lists for FMcKE relevant sources
E_res_p=[]
E_res_g=[]
f_res_p=[]
f_res_g=[]
E_res_tot=[]
#total energy list
E_tot=[]
#evolved orbital energy
correct_orbital_energy=[]
#forcing freq
force_freq=[]
#res conds
hires=[]
lowres=[]
#
delres=high_end-w_force
del_res_cond=[]
#
pmode=[]
gmode=[]
orbyengy=[]
```

```python

#bin counter
n=1
#append intial values into lists
for i in range(1):
    #normal stuff
    semi_major.append(a0)
    newecc.append(ecc)
    orbital_period.append(porb0)
    frequency.append(freq0)
    #fmcke stuff
    E_res_p.append(Ep0)
    E_res_g.append(Eg0)
    Grav_wave_energy.append(E_GW)
    E_res_tot.append(Ert0)
    ###
    #E_tot.append(Etotal0)
    ####
    #changed orbital energy
    correct_orbital_energy.append(eorbtot)
    #record forcing frequency
    force_freq.append(w_force)
    hires.append(high_end)
    lowres.append(low_end)
    del_res_cond.append(delres)
    #mode check
    pmode.append(pmode0)
    gmode.append(gmode0)
    orbyengy.append(Eorb)
#begin the dual evolution loop
for i in range(1,len(cosmicpop)):
    while n<=200:
        timestep_new=n*t_gw/100
        timestep_old=(n-1)*t_gw/100
        
        #define time step for bins
        #dt=timestep_new-timestep_old
        
        dt=porb
        #fchirp=(96/5)*((c**3)/G)*(f/mc)*((G/c**3)*pi*f*mc)**(8/3)
        
        
        
        #begin FMcKE if resonant condition is met
        while np.any(low_end<=w_force & w_force<=high_end):
            k=0
            r=0
            #is the resonant mode a gravity mode?
            while np.any(w_force==(d_res_mode_g*w_star1)):
                #truncate mode if harmonic/quantum# is too high
                if k==10 or q==10:
                    break

                #freqency of mode
                w_jg=d_res_mode_g*w_star1
                #overlap term for overlapping resonance
                X_jg=27*(w_jg/w_star1)**3.7
                #Satured g mode terms
                w_force_dotg=w_jg-(1/(r*2*tdg**2))
                t_forceg=1/((w_force_dotg)**1/2)


                #is the mode saturated?
                if t_forceg>tdg:
                    #energy deposited
                    Eg=(10**-5)*Estar_1*((w_jg/w_star1)/0.1)**(-2/3)*(X_jg/10**-3)*((mc/mass_1Solar)**(4/5))*((a_b/mass_1rSolar)/10)**-4
                    E_res_g.append(Eg)
                #if not saturated use unsaturated energy
                else:
                    Eg=((10**-5)*Estar_1*((w_jg/w_star1)/0.1)**(-2/3)*(X_jg/10**-3)*((mc/mass_1Solar)**(4/5))*((a_b/mass_1rSolar)/10)**-4)*(10**(-6)*(t_forceg/0.115)**(2/3)*(tdg/(115.74*10**6))**(-2/3))
                    E_res_g.append(Eg)
                
                
                #cooling time of mode via thermal process
                t_j_ktg=(((4*pi)/w_j1)*(Estar_1/Eg)**(1/2))
                #cooling time of mode via grav wave
                t_j_GWg=((5*c**5)/G)*(1/(Eg*w_j1**2))
                #decay time of mode
                tdg=t_j_ktg+t_j_GWg
                w_force1+=w_force_dot1*dt
                
                
                #update mode
                d_res_mode_g+=0.1
                r+=1
                #harmonic count up
                k+=1
                
                
                #update resonant conditions
                low_end=w_star1/k-1/tdg
                high_end=w_star1/k+1/tdg
            
            
            #is the resonant mode a pressure mode?
            while np.any(w_force==(d_res_mode_p*w_star1)):
                #truncate
                if k==10 or q==10:
                    break
                
                w_jp=d_res_mode_p*w_star1
                X_jp=27*(w_jp/w_star1)**3.7
                Np=5
                
                
                #Satured p_mode
                w_force_dotp=w_jp-(1/(r*2*tdp**2))
                t_forcep=1/((w_force_dotp)**1/2)
                if t_force1>tdp:
                    Ep=0.1*Estar_1*((w_star1/(3.4722*10**(-6)))**(4/5))*(((w_jp/wstar_1))/4)**(2/5)*((k/1)**(-2))*((mc/mass_1Solar)**(4/5))*(((a_b/mass_1rSolar)/10)**(-12/5))*((X_jp/(10**(-3)))**(3/5))*((t_p/115.74)**(4/5))*((fchirp/0.5)**(-4/5))*(((Np/10**2))**(-4/5))
                    E_res_p.append(Ep)
                #unsaturated energy
                else:
                    Ep=0.1*Estar_1*((w_star1/(3.4722*10**(-6)))**(4/5))*(((w_jp/w_star1))/4)**(2/5)*((k/1)**(-2))*((mc/mass_1Solar)**(4/5))*(((a_b/mass_1rSolar)/10)**(-12/5))*((X_jp/(10**(-3)))**(3/5))*((t_p/(115.74)**(4/5)))*((fchirp/0.5)**(-4/5))*(((Np/10**2))**(-4/5))*(0.06*(t_forcep/0.1157)**(2/5)*(tdp/115.74)**(-2/5))
                    E_res_p.append(Ep)
                
                #decay time of pmode
                t_j_ktp=2*(t_p/Np)*(Estar_1/E1)
                t_j_GWp=((5*c**5)/G)*(1/(E1*w_j1**2))
                tdp=t_j_ktp+t_j+GWp
                
                #harmonic counter
                k+=1
                r+=1
                d_res_mode_p+=1
                
                #update resonant conditions
                
                low_end=w_star1/k-1/tdp
                high_end=w_star1/k+1/tdp
        
        #change eccentricity
        decc=(-19/12)*(Beta/cn**4)*((ecc**(-29/19))*(1-ecc**2)**(3/2))/((1+(121/304)*ecc**2)**(1181/2299))
        ecc+=decc*dt
        newecc.append(ecc)
        
        
        #calc total energy of binary
        #gravitational wave energy
        dE_GW=-(32/5)*(((G**4*mass_1Solar**2*mass_2Solar**2)*(mass_1Solar+mass_2Solar))/(c**5*a_b**5*(1-ecc**2)**(7/2)))*(1+(73/24)*ecc**2+(37/96)*ecc**4)
        E_GW+=dE_GW*dt
        Grav_wave_energy.append(E_GW)
        
        #energy deposited into resonant mode
        Ert=Eg+Ep
        E_res_tot.append(Ert)
        
        #orbital energy
        Eorb=-(G*(mass_1Solar+mass_2Solar))/(2*a_b)
        orbyengy.append(Eorb)
        
        #new orbital energy accounting for GWave Emission/Resonant deposits
        if Ert !=0:
            eorbtot=Eorb+E_GW+Ert
        else: eorbtot=Eorb+E_GW
        correct_orbital_energy.append(eorbtot)
        
        #calculate semi-major using total energy or Kepler III
        a_b=(-G*mass_1Solar*mass_2Solar)/(2*eorbtot)
        #a_b=((porb**2*G*mass_1Solar*mass_2Solar)/(4*pi**2*mu))**(1/3)
        semi_major.append(a_b)
        
        
        #orbital period using energy or Kepler III
        porb=(4*pi**2*a_b**3/(G*(mass_1Solar+mass_2Solar)))**(1/2)
        #porb=pi*G*mass_1Solar*mass_2Solar*(mu/2)**(1/2)*(abs(Etotal)**(-3/2))
        orbital_period.append(porb)
        
        
        f=2/porb
        frequency.append(f)

        cn=a_b*(1-ecc**2)*(ecc**(-12/19))*(1+(121/304)*ecc**2)**(-870/2299)
        
        #pricipal number change
        #q=(1/2**(1/2))*(w_force/(k*w_star1)+2*f_spin1)*(a_b/mass_1rSolar)**(3/2)
        #update break up spin fraction
        f_spin1+=.1
        #spin of WD
        w_spin1=f_spin1*w_star1
        #forcing frequency of WD
        w_force=q*porb-m*w_spin1
        force_freq.append(w_force)
        
        hires.append(high_end)
        lowres.append(low_end)
        
        #how far away from res
        delres=high_end-w_force
        del_res_cond.append(delres)
        
        ppmode=d_res_mode_p*w_star1
        ggmode=d_res_mode_g*w_star1
        pmode.append(ppmode)
        gmode.append(ggmode)
        n=n+1
        if np.all(r_l_tot>=a_b):
            break
####################
```
