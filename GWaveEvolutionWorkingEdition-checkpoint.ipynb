```python
import pandas as pd
from math import *
from scipy import interpolate
import matplotlib.pyplot as plt
import numpy as np
import decimal

#read in data from cosmic using path to file
cosmicpop = pd.read_csv(r"PathToFile", index_col=False)
PorbSI=cosmicpop.PorbSI
mass_1SI=cosmicpop.mass_1SI
mass_2SI=cosmicpop.mass_2SI
mass_1rSI=cosmicpop.mass_1rSI
mass_2rSI=cosmicpop.mass_2rSI
a_b=cosmicpop.a_b
ecc=cosmicpop.ecc
mc=cosmicpop.mc

#re convert values for sensitivity
############
# PorbDay=PorbSI/86400
# #PorbYear=porb/356.256
# #PorbDet=detectable
# #convert mass1/2 from solar mass to kg
# mass_1Solar=mass_1SI/1.98847e30
# mass_2Solar=mass_2SI/1.98847e30
# #convert radii to m
# mass_1rSolar=mass_1rSI/6.96e8
# mass_2rSolar=mass_2rSI/6.96e8
# #chirp mass
# #drop values that are 0
# mc=((mass_1Solar*mass_2Solar)**(3/5)/((mass_1Solar+mass_2Solar)**(1/5)))
# #semi-major for R solar to m
# a_b=a_b/6.96e8
# w_star1=cosmicpop.w_star1
# w_star2=cosmicpop.w_star2
# #angular velocity from s^-1 to yr^-1
# w_star1=w_star1/86400
# w_star2=w_star2/86400

#constants
#gravitational converted from AU**3*yr**-2*Msolar**-1 to Solar Radii*day**-2*Msolar**-1
G=3100670.345563776
#speed of light in solar radii per day
c=37231.7
#define roche lobe radii
q_1=mass_1Solar/mass_2Solar
r_l_1=(0.49*q_1**(2/3))/(0.6*q_1**(2/3)+np.log(1+q_1**(1/3)))
q_2=mass_2Solar/mass_1Solar
r_l_2=(0.49*q_2**(2/3))/(0.6*q_2**(2/3)+np.log(1+q_2**(1/3)))
r_l_tot=r_l_1+r_l_2
#reduced mass
mu=(mass_1Solar*mass_2Solar)/(mass_1Solar+mass_2Solar)
#times
#t=0
#tobs=1.262*10**8
t_gw=(5/128)*((c**5)/(G**3))*((a_b**4)/((mass_1Solar+mass_2Solar)**2*mu))*(1-ecc**2)**(7/2)
#Constants required for gwave evolution
Beta=(64/5)*c**(-5)*G**3*mass_1Solar*mass_2Solar*(mass_1Solar+mass_2Solar)

a0=a_b

c0=a_b*(1-ecc**2)*ecc**(-12/19)*(1+(121/304)*ecc**2)**(-870/2299)

cn=a_b*(1-ecc**2)*ecc**(-12/19)*(1+(121/304)*ecc**2)**(-870/2299)

porb0=(a0**3/(mass_1Solar+mass_2Solar))**(1/2)

porb=(a_b**3/(mass_1Solar+mass_2Solar))**(1/2)

freq0=2/porb0

Eorb=-(G*(mass_1Solar+mass_2Solar))/(2*a_b)

E_GW=0

eorbtot=Eorb+E_GW

n=1
#evolve values
semi=[]
newecc=[]
con=[]
NewOrb=[]
orbitalperiod=[]
orbitalfrequency=[]
chirpmass=[]
mass1=[]
mass2=[]
mass1r=[]
mass2r=[]
wstar1=[]
wstar2=[]
for i in range(1):
    con.append(c0)
    semi.append(a0)
    newecc.append(ecc)
    orbitalperiod.append(porb0)
    orbitalfrequency.append(freq0)
    NewOrb.append(eorbtot)
    chirpmass.append(mc)
    mass1.append(mass_1Solar)
    mass2.append(mass_2Solar)
    mass1r.append(mass_1rSolar)
    mass2r.append(mass_2rSolar)
    wstar1.append(w_star1)
    wstar2.append(w_star2)


for i in range(1,len(cosmicpop)):
    while n<=100:
        timestep_new=n*t_gw/100
        timestep_old=(n-1)*t_gw/100

        #dt=timestep_new-timestep_old
        dt=porb
        
        decc=(-19/12)*(Beta/cn**4)*((ecc**(-29/19))*(1-ecc**2)**(3/2))/((1+(121/304)*ecc**2)**(1181/2299))
        ecc+=decc*dt
        newecc.append(ecc)
        
        #gwave energy
        dE_GW=-(32/5)*(((G**4*mass_1Solar**2*mass_2Solar**2)*(mass_1Solar+mass_2Solar))/(c**5*a_b**5*(1-ecc**2)**(7/2)))*(1+(73/24)*ecc**2+(37/96)*ecc**4)
        E_GW+=dE_GW*dt
        
        Eorb=-(G*(mass_1Solar+mass_2Solar))/(2*a_b)
        
        eorbtot=Eorb+E_GW
        
        #a_b=((cn*ecc**(12/19))/(1-ecc**2))*(1+(121/304)*ecc**2)**(870/2299)
        a_b=(-G*mass_1Solar*mass_2Solar)/(2*eorbtot)
        semi.append(a_b)

        porb=(a_b**3/(mass_1Solar+mass_2Solar))**(1/2)
        orbitalperiod.append(porb)

        fre=2/porb
        orbitalfrequency.append(fre)

        cn=a_b*(1-ecc**2)*(ecc**(-12/19))*(1+(121/304)*ecc**2)**(-870/2299)
        con.append(cn)
        chirpmass.append(mc)
        mass1.append(mass_1Solar)
        mass2.append(mass_2Solar)
        mass1r.append(mass_1rSolar)
        mass2r.append(mass_2rSolar)
        wstar1.append(w_star1)
        wstar2.append(w_star2)
        n=n+1
    if np.any(r_l_tot>=a_b):
        break
        
#zip files to flatten
mesemi=list(zip(*semi))
meorb=list(zip(*orbitalperiod))
mefre=list(zip(*orbitalfrequency))
mechirp=list(zip(*chirpmass))
mecc=list(zip(*newecc))

data=[mesemi, meorb, mefre,mechirp, mass1, mass2, mass1r, mass2r, wstar1, wstar2, mecc]
#flatten list of lists into list
x=[item for sublist in mesemi for item in sublist]
y=[item for sublist in meorb for item in sublist]
z=[item for sublist in mefre for item in sublist]
xx=[item for sublist in mechirp for item in sublist]
yy=[item for sublist in mass1 for item in sublist]
zz=[item for sublist in mass2 for item in sublist]
xxx=[item for sublist in mass1r for item in sublist]
yyy=[item for sublist in mass2r for item in sublist]
zzz=[item for sublist in wstar1 for item in sublist]
xyz=[item for sublist in wstar2 for item in sublist]
zzzz=[item for sublist in mecc for item in sublist]

data=[x,y,z,xx, yy, zz, xxx, yyy, zzz, xyz, zzzz]
#create new dataframe
df=pd.DataFrame(data)
df=df.transpose()
df=df.rename(columns={0: 'SemiMajorAxis', 1: 'OrbitalPeriod', 2: 'OrbitalFrequency', 3:'Chirpmass', 4:'mass1',5:'mass2',6:'mass1r',7:'mass2r',8:'wstar1',9:'wstar2', 10:'ecc'})


#read in values from new data frame
sma=df.SemiMajorAxis
sma.replace(np.nan,0)
op=df.OrbitalPeriod
op.replace(np.nan,0)
of=df.OrbitalFrequency
of.replace(np.nan, 0)
cm=df.Chirpmass
mass1=df.mass1
mass2=df.mass2
mass1r=df.mass1r
mass2r=df.mass2r
wstar1=df.wstar1
wstar2=df.wstar2
eccen=df.ecc
eccen.replace(np.nan,0)

#create new data frame for export
newdata=[sma, op, of, cm, mass1, mass2, mass1r, mass2r, wstar1, wstar2,eccen]
newdf=pd.DataFrame(newdata)
newdf=newdf.transpose()
newdf=newdf.rename(columns={'sma': 'SemiMajorAxis', 'op': 'OrbitalPeriod', 'of': 'OrbitalFrequency', 'cm':'Chirpmass', 'eccen':"eccentricity"})

newdf.to_csv(r"PathToFile")
```
